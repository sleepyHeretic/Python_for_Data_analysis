```python
# 7.
df_low = df_HR.loc[df_HR['salary'] == 'low']
df_high = df_HR.loc[df_HR['salary'] == 'high']

df_low
```

```python
sns.scatterplot(df_low['average_montly_hours'])
```

```python
sns.scatterplot(df_high['average_montly_hours'])
```

```python
# Н0: среднее время на работе у сотрудников с высоким окладом равно среднему времени на работе сотрудников с низким окладом
# Н1: показатели времени проведенного на работе у сотрудников не равны

stat, p_value = st.ttest_ind(df_high['average_montly_hours'],df_low['average_montly_hours'])

if p_value < 0.05:
  print(f"{p_value} - Отклоняем нулевую гипотезу. Средние вероятно различаются.")
else: print(f"{p_value} - Не отклоняем нулевую гипотезу. Средние, вероятно, равны.")
```

```python
# 8.
df_quitted = df_HR.loc[df_HR['left'] == 1]
df_employee = df_HR.loc[df_HR['left'] == 0]

# показатели :
# Доля сотрудников с повышением за последние 5 лет
# Средняя степень удовлетворенности
# Среднее количество проектов

quitted_prom = len(df_quitted.loc[df_quitted['promotion_last_5years'] == 1]) * 100 / len(df_quitted)
employee_prom = len(df_employee.loc[df_employee['promotion_last_5years'] == 1]) * 100 / len(df_employee)

dict_employee = {'Доля сотрудников с повышением за последние 5 лет, в %': [quitted_prom, employee_prom],
                 'Средняя степень удовлетворенности': [df_quitted['satisfaction_level'].median(), df_employee['satisfaction_level'].median()],
                 'Среднее количество проектов': [df_quitted['number_project'].median(), df_employee['number_project'].median()]
                 }
df_stat = pd.DataFrame.from_dict(dict_employee, orient = "index",columns = ['уволившиеся','не уволившиеся'])
# показатели сотрудников:
```

```python
# 9.
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
data_class = pd.array(df_HR["left"])

# разделяем выборки:
X_train,X_test, y_train, y_test = train_test_split(df_HR[['Work_accident','satisfaction_level','last_evaluation','number_project',
                                                         'average_montly_hours','time_spend_company','Work_accident',
                                                         'promotion_last_5years']], data_class)
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
lda.fit(X_train,y_train)
```

```python
result = pd.DataFrame(y_test, lda.predict(X_test)).T
result
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,lda.predict(X_test)))
# Точность модели довольно низка. От прогнозирования увольнения сотрудника на основании этой модели стоит отказаться
```
